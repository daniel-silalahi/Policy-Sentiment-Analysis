Create the Cloropleth Map

```python
# import libraries
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd

# read in data
url = 'https://raw.githubusercontent.com/Allen-Bian/QM2-Final-project/main/data.csv'
df = pd.read_csv(url)

# read in shapefile
df_map = gpd.read_file('cb_2021_us_state_20m.shp')

# merge dataframes. name of state in df_map is 'NAME' whilst in df is 'STATE'.
merged = df_map.set_index('NAME').join(df.set_index('STATE'))

# create list of columns (everything except state)
columns = df.columns[1:]

for column in columns:
  # create figure and axes for Matplotlib
  fig, ax = plt.subplots(1, figsize=(50, 10))
  # create map
  merged.plot(column=column, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
  # turn axis off to remove ticks and labels
  ax.axis('off')
  ax.set_title(column, fontdict={'fontsize': '20', 'fontweight' : '3'})
  plt.show()
```
